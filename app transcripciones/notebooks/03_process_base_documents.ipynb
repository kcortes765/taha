# 03 Process Base Documents

Extrae texto y fragmenta documentos base desde los manifests.

```python
from cis.domain.models import ExtractedDocument, BaseDocumentFragment
from cis.config.settings import get_settings
from cis.storage.manifest_store import ManifestStore
from cis.storage.paths import CISPaths

paths = CISPaths(get_settings())
docs = ManifestStore(paths.documents_manifest('sismo'), ExtractedDocument).all()
fragments = ManifestStore(paths.fragments_manifest('sismo'), BaseDocumentFragment).all()
len(docs), len(fragments)
```
